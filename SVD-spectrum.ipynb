```python
import numpy as np
import matplotlib.pyplot as plt
import pycolate
from scipy.linalg import svd
```

```python
size = 1000
probabilities_list = np.linspace(0.0000001,1,50)
```

```python
percolated_list = []
clustersize_list = []
spectrum = []
for p in probabilities_list:
    print('p = ', p)
    Perc = pycolate.Percolation(size, size, p)
    Perc.cluster_find()
    percolated_list.append(Perc.percolated)
    clustersize_list.append(Perc.mean_cluster_size)
    spectrum.append(svd(Perc.config, compute_uv=False))
    del Perc
```

```python
plt.plot(probabilities_list, np.array(spectrum)[:, 1:10])
# plt.ylim(0,35)
plt.show()
```
