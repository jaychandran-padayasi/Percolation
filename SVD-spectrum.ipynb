```python
import numpy as np
import matplotlib.pyplot as plt
import pycolate
from scipy.linalg import svd
```

```python
size = 200
probabilities_list = np.linspace(0.0000001,1,50)
```

```python
percolated_list = []
clustersize_list = []
spectrum = []
for p in probabilities_list:
    print('p = ', p)
    temp_perc = []
    temp_clust = []
    for index in range(200):
        Perc = pycolate.Percolation(size, size, p)
        Perc.cluster_find()
        temp_perc.append(int(Perc.percolated))
        temp_clust.append(Perc.mean_cluster_size)
    # spectrum.append(svd(Perc.config, compute_uv=False))
        del Perc
    percolated_list.append(np.average(temp_perc))
    clustersize_list.append(np.average(temp_clust))
```

```python
plt.plot(probabilities_list, np.sqrt(np.array(spectrum))[:, 1:])
# plt.xlim(0.45,0.55)
# plt.ylim(0, 35)
plt.show()
```

```python
plt.plot(probabilities_list, percolated_list, '-o')
```

```python
plt.plot(probabilities_list, clustersize_list)
```
